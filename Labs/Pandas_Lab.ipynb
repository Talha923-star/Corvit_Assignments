```python
# Create a Pandas Series from a list of numbers 10 to 50
import pandas as pd
list = range(10,50)
Series = pd.Series(list)
print(Series)
```

```python
# Create a Series with custom index: ['a','b','c','d'].
import pandas as pd
Series = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
print("Series with Custom Index")
print(Series)
```

```python
# Find the mean, max, and min of a Series
import pandas as pd
data = pd.Series([10, 20, 30, 40, 50])
print("Mean of Series is:",data.mean())
print("Max of Series is:",data.max())
print("Min of Series is:",data.min())
```

Create a DataFrame from a dictionary:
{
  "Name": ["Ali", "Sara", "Ahmed"],
  "Age": [22, 25, 23],
  "Marks": [85, 90, 88]
}

```python
# Get shape of DataFrame
import pandas as pd
Student = {
     "Name": ["Ali", "Sara", "Ahmed"],
     "Age": [22, 25, 23],
     "Marks": [85, 90, 88]
     }

df = pd.DataFrame(Student)
print(df)
print("Shape of DataFrame is:",df.shape)
```

```python
# Select only the Name column
import pandas as pd
Student = {
     "Name": ["Ali", "Sara", "Ahmed"],
     "Age": [22, 25, 23],
     "Marks": [85, 90, 88]
     }

df = pd.DataFrame(Student)
print("Name Column is: \n",df['Name'])
```

```python
# Select multiple columns (Name and Marks).
import pandas as pd
Student = {
     "Name": ["Ali", "Sara", "Ahmed"],
     "Age": [22, 25, 23],
     "Marks": [85, 90, 88]
     }

df = pd.DataFrame(Student)
print("Name and Marks of Student is: \n", df[['Name','Marks']])
```

```python
# Select rows where Age > 23.
import pandas as pd
Student = {
     "Name": ["Ali", "Sara", "Ahmed"],
     "Age": [22, 25, 23],
     "Marks": [85, 90, 88]
     }

df = pd.DataFrame(Student)
print("Rows where Age > 23 is: \n",df[df['Age']>23])
```

```python
# Select students with Marks greater than 85
import pandas as pd
Student = {
     "Name": ["Ali", "Sara", "Ahmed"],
     "Age": [22, 25, 23],
     "Marks": [85, 90, 88]
     }

df = pd.DataFrame(Student)
print("Rows where Marks > 85 is: \n",df[df['Marks']>85])
```
