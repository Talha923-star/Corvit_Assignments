```python
correct_pin = 12354
balance = 1000

def ent_pin():
    entered_pin = int(input("Enter Your Pin: "))
    if entered_pin == correct_pin:
        return True
    else:
        print("Incorrect Pin!!!")
        return False


def check_balance():
    global balance
    print("Your Current Balance is:", balance)


def deposit_cash():
    global balance
    amount = int(input("Enter the amount you want to deposit: "))
    if amount > 0:
        balance += amount
        print(amount, "Deposited Successfully")
        print("New Balance is:", balance)
    else:
        print("Invalid Input")


def withdrawal_cash():
    global balance
    amount = int(input("Enter the amount you want to withdraw: "))
    if amount > 0:
        if amount <= balance:
            balance -= amount
            print(amount, "Withdrawal Completed")
            print("Remaining Balance is:", balance)
        else:
            print("Insufficient Balance!!!")
    else:
        print("Amount Must be positive")


def atm_system():
    if ent_pin():
        while True:
            print("\nWelcome to ATM")
            print("1. Check Balance")
            print("2. Deposit Cash")
            print("3. Withdrawal Cash")
            print("4. Exit")

            choice = int(input("Enter a choice between 1-4: "))

            if choice == 1:
                check_balance()
            elif choice == 2:
                deposit_cash()
            elif choice == 3:
                withdrawal_cash()
            elif choice == 4:
                print("Thanks For Using ATM, Bye!")
                break
            else:
                print("Invalid Input! Choose From 1-4")


atm_system()
```
