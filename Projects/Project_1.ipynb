5. Check and Handle Null Values
  * Check how many null values exist in each column.
    * Decide whether to:
    * Remove rows containing null values, or
    * Fill them using suitable methods (mean, median, mode, etc.).
* Apply the method you choose.

```python
# Check how many null values exist in each column.
print(df.isnull().sum())
# Decide whether to:
# Mean for Gdp,Alcohol,etc  , # Mediam for Population
df['Life expectancy '] = df['Life expectancy '].fillna(df['Life expectancy '].mean())
df['Alcohol'] = df['Alcohol'].fillna(df['Alcohol'].mean())
df['Hepatitis B'] = df['Hepatitis B'].fillna(df['Hepatitis B'].mean())
df['BMI '] = df['BMI '].fillna(df['BMI '].mean())
df['Total expenditure'] = df['Total expenditure'].fillna(df['Total expenditure'].mean())
df['GDP'] = df['GDP'].fillna(df['GDP'].mean())
df['Population'] = df['Population'].fillna(df['Population'].median())
df['Schooling'] = df['Schooling'].fillna(df['Schooling'].mean())
```

8. Explore Categorical Data
   * Use groupby() to analyze categorical columns.
   * Find counts, averages, or totals based on categories.

```python
# group by status
status_analysis = df.groupby("Status")["Life expectancy"].mean()
print("Average Life Expectancy by Status:")
print(status_analysis)
# group by country
country_analysis = df.groupby("Country")["Life expectancy"].mean()
print("\nAverage Life Expectancy by Country:")
print(country_analysis.head())
# multiple column analysis
multi_analysis = df.groupby("Status")[["Life expectancy","GDP","Schooling"]].mean()
print("\nMultiple Column Analysis:")
print(multi_analysis)
```

10. Trend Analysis Using Visualization

Bar Chart between life expectancy by Status
* Show Status and Life expectancy
* Show Number of people in developing or developed country


```python
import matplotlib.pyplot as plt
import seaborn as sns
# Bar Chart between life expectancy by Status
life_status = df.groupby('Status')['Population'].sum()
plt.bar(life_status.index,life_status.values, color='skyblue', edgecolor='black')
plt.title('Population by Status')
plt.xlabel('Status')
plt.ylabel('Population')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Line Chart between life expectancy by years
* Show Life expectancy by Years in Different Countries
* Show Years and Age of people in that Year

```python
# Line Chart between life expectancy by years
plt.plot(df['Life expectancy'],df['Year'] , marker='o')
plt.title('Life Expectancy over Years')
plt.xlabel('Life expectancy')
plt.ylabel('Year')
plt.grid(True)
plt.show()
```

Histogram of life expectancy

```python
# Histogram of life expectancy
plt.hist(df['Life expectancy'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Life expectancy')
plt.title('histogram of Life expectancy')
plt.show()
```

Scatter Plot  bwt Life expectancy,GDP,Schooling.

*  Show Life Expectancy,gdp,schooling

Box Plot Life Expectancy distribution by Status
* Show life expectancy according to Status
* Show People of different Age is Living in Country

```python
# Box Plot Life Expectancy distribution by Status
plt.figure(figsize=(12, 6))
sns.boxplot(x='Status', y='Life expectancy', data=df, palette='Set2')
plt.title('Life Expectancy Distribution by Status')
plt.xlabel('Status')
plt.ylabel('Life Expectancy')
plt.show()
```

 Pie Chart

```python
# Pie Chart  of Status distribution
status_counts = df['Status'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#ff9999'])
plt.title('Distribution of Countries by Status')
plt.show()
```
