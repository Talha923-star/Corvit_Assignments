```python
'''1.Write a program that calculates the electricity bill based on units consumed:
•	First 100 units → 5 per unit
•	Next 100 units → 8 per unit
•	Above 200 units → 10 per unit'''
units = int(input("Enter the units of your Bill: "))
bill_units = 0
if (units <= 100):
    bill_units = units * 5
elif (units <= 200):
    bill_units = (100*5) + ((units - 100)*8)
else:
    bill_units = (100*5)+(100*8)+((units-200)*10)
print(bill_units)
```

```python
'''2. Write a program that takes a student's marks as input and prints the grade:
•	90 and above → A
•	75–89 → B
•	50–74 → C
•	Below 50 → Fail'''
marks = int(input("Enter the Marks to Find the Grade: "))
if marks >= 90 :
    print("Your Grade is A")
elif 75<= marks and marks <=89:
    print("Your Grade is B")
elif 50<= marks and marks <=74:
    print("Your Grade is C")
else:
    print("You are Fail")
```
