```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
file_path = '/content/drive/MyDrive/Student Dataset.csv' #To load Data From Drive
df = pd.read_csv(file_path)
print("File load Successfully")
print("First 5 values of the File: \n",df.head(5)) #Printing First 5 Values
print("The Information Of File: \n",df.info()) #printing info of Dataset
```

```python
# Part 2 — Null Analysis 4. Find the number of null values in each column.
#5. Create a new column called Null_Count that contains the number of null values in each row
df = pd.read_csv(file_path)
nul = df.isnull().sum()
print("Total Null Values \n",nul)

df['Null_Count'] = df.isnull().sum(axis=1)
print("Null Vaues in All Colum \n", df['Null_Count'])
```

```python
# Part 3 — Handling Null Values
# 6. Identify the column that contains only null values and remove it from the dataset.
# 7. Handle the missing values present in the dataset using appropriate techniques.
df = df.dropna(axis=1, how='all')
print("Drop columns with all missing values")
print(df)
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['ProjectScore'].fillna(df['ProjectScore'].bfill(), inplace=True)
df['Attendance'].fillna(df['Attendance'].median(), inplace=True)
df['Marks'].fillna(df['Marks'].mean(), inplace=True)
print(df.isnull().sum())
```

```python
''' Part 4 — Conditional Filtering
Display rows that satisfy the following conditions:
8. Students with Marks greater than 85.
9. Students from Lahore whose Attendance is greater than 85.
10. Students whose ProjectScore is greater than 85 and Marks greater than 80.'''
print(df[df['Marks'] > 85])
print(df[(df['City'] == 'Lahore') & (df['Attendance'] > 85)])
print(df[(df['ProjectScore'] > 85) & (df['Marks'] > 80)])
```

```python
'''Part 5 — Data Manipulation
11. Increase the Marks of all students by 5.
12. Increase Marks by 5 only for students whose Marks are less than 80.
13. Create a new column named TotalScore which is the sum of Marks and ProjectScore.
14. Create another column named PerformanceIndex which is the product of Marks and
Attendance.'''

df['Marks'] = df['Marks'] + 5

df.loc[df['Marks'] < 80, 'Marks'] += 5

df['TotalScore'] = df['Marks'] + df['ProjectScore']

df['PerformanceIndex'] = df['Marks'] * df['Attendance']

print(df.head())
```

```python
'''Part 6 — GroupBy Operations
15. Find the average Marks for each Department.
16. Find the number of students in each City.'''
Average_Marks = df.groupby('Department') ['Marks'].mean()
print(Average_Marks)
City_Count = df['City'].value_counts()
print(City_Count)
```

```python
'''Part 7 — GroupBy with Aggregation
17. For each Department, calculate:
 Average Marks
 Maximum Marks
 Minimum Marks
18. For each City, calculate:
 Average Attendance
 Maximum ProjectScore'''
Department_Stats = df.groupby('Department')['Marks'].agg(['mean','max','min'])
print(Department_Stats)
City_Stats = df.groupby('City').agg({'Attendance':'mean','ProjectScore':'max'})
print(City_Stats)
```
